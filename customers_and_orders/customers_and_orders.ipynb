```python
from os.path import exists
import csv

# check if file can be found and open the file
datafile = "./sample_data/customers.csv"
if not exists(datafile):
  raise SystemExit("You should run the first code cell and download the dataset files!")

# customers file is in the sample_data folder and can be opened.
# Using csv reader is one way of dealing with the file
# but it's up to you how you want to proceed
with open(datafile, 'r') as fl:
  csvreader = csv.reader(fl, delimiter=',')

  # loop through the rows and count etc. THE REST IS UP TO YOU!

def read_customers(filepath):
    with open(filepath, 'r') as f:
        reader = csv.reader(f, delimiter=',')
        headers = next(reader)  # encabezados
        data = [row for row in reader]  # todas las filas

    return headers, data

print("All done!")
```

```python
headers, customers = read_customers(datafile)

print("Headers:", headers)
print("Primeras filas:", customers[:5])
```

```python
#muestra con un show el datafram del csv de customer
import csv

def count_customers(filepath):
    with open(filepath, 'r') as f:
        reader = csv.reader(f, delimiter=',')
        next(reader)  # saltar el header
        count = sum(1 for _ in reader)
    return count

# uso
total_customers = count_customers(datafile)
print(f"Total de clientes: {total_customers}")
```

```python
#Question #4: What is the number of orders placed in October 2021?
import pandas as pd

df = read_data(
    "./sample_data/orders.csv"
    ,columns=["OrderID","Date"]
)

# Convert 'Date' column to datetime objects, coercing errors to NaT
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Filter for orders in October 2021, automatically excluding NaT values
df_october_orders = df[
    (df['Date'].dt.year == 2021) &
    (df['Date'].dt.month == 10)
]

# Count the number of unique orders in October 2021
num_orders_october = df_october_orders['OrderID'].nunique()
print(f"Number of orders placed in October 2021: {num_orders_october}")
```

```python
#Question #5: Which customer spent the most amount of money in 2021?
import pandas as pd

df = read_data(
    "./sample_data/orders.csv"
    ,columns=["CustomerID","Date","Price"]
)

# Convert 'Date' column to datetime objects, coercing errors to NaT
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Filter for orders in 2021, automatically excluding NaT values
df_2021 = df[
    (df['Date'].dt.year == 2021)
]

# Group by CustomerID and sum OrderTotal to find total spent by each customer
df_customer_grouped = df_2021.groupby(['CustomerID']).agg({'Price': 'sum'}).reset_index()

# Sort by OrderTotal in descending order to find the customer who spent the most
df_customer_grouped_desc = df_customer_grouped.sort_values('Price', ascending=False)

# Print the customer who spent the most
print("Customer who spent the most in 2021:")
print(df_customer_grouped_desc.head(1))
```

```python
#Question #6: Historically, what is the best month for sales?
df = read_data(
    "./sample_data/orders.csv"
    ,columns=["Date","OrderID","OrderTotal"]
)

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Date'] = df['Date'].dt.to_period('M')

# Group by 'Date' (month) and count orders, then sort by OrderID count in descending order
order_history_grouped_quantity = df[['Date', 'OrderID']].groupby('Date').count().reset_index().sort_values(by='OrderID', ascending=False)

pd.set_option('display.max_rows', None)
print(order_history_grouped_quantity.head(1))


order_history_grouped_sum = df[['Date', 'OrderTotal']].groupby('Date').sum("OrderTotal").reset_index().sort_values(by='OrderTotal', ascending=False)
print(order_history_grouped_sum.head(1))
```

```python
#Question #6: Historically, what is the best month for sales?

df_pro = df.assign(
    year=df['Date'].dt.year,
    month=df['Date'].dt.month
)

# Group by month (across all years) and calculate the average OrderTotal
monthly_avg_sales = df_pro.groupby('month').agg({'OrderTotal': 'mean'}).reset_index()

# Sort by the average OrderTotal in descending order to find the best month
best_month_for_sales = monthly_avg_sales.sort_values(by='OrderTotal', ascending=False)

print("Month with the highest average sales (historically):")
print(best_month_for_sales.head(1))

# Optionally, you can map month numbers to names for better readability
month_names = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}
best_month_for_sales['MonthName'] = best_month_for_sales['month'].map(month_names)
print("\nMonth with the highest average sales (with names):")
print(best_month_for_sales.head(1))
```
